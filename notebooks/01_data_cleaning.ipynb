```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/Algerian_forest_fires_dataset.csv")
df.head()
```

```python
print("Dataset shape:",df.shape)
print("-" * 50)
df.info()
```

```python
# Create a fresh copy to work safely
df_clean = df.copy()

# 1. Clean up whitespace from the column names to avoid hidden space issues
df_clean.columns = df_clean.columns.str.strip()

# 2. Assign Regions based on your exact layout
# Everything up to index 121 is Bejaia (0)
df_clean.loc[:121, "Region"] = 0

# Everything from index 122 onwards is Sidi-Bel Abbes (1)
df_clean.loc[122:, "Region"] = 1

# 3. DROP the garbage rows FIRST (this removes the NaN values that caused the error)
# Drop any row where the 'day' column is completely missing/NaN
df_clean = df_clean.dropna(subset=['day']).reset_index(drop=True)

# Drop the row where 'day' is literally written out as the text string "day"
df_clean = df_clean[df_clean['day'] != 'day'].reset_index(drop=True)

# 4. NOW it is safe to convert Region into a clean integer type!
df_clean["Region"] = df_clean["Region"].astype(int)

# Let's check the shape and verify the boundary point transitions beautifully
print("Cleaned Dataset Shape:", df_clean.shape)
df_clean.iloc[120:125]
```

```python
# Remove the stray region title row sitting at index 122
df_clean = df_clean[df_clean['day'] != 'Sidi-Bel Abbes Region Dataset'].reset_index(drop=True)

# Let's inspect the boundary rows one last time to ensure it's 100% flawless
df_clean.iloc[120:124]
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df = df[df['day'].str.strip() != 'day'].reset_index(drop=True)
```

```python
for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype == 'str':
        df[col] = df[col].str.strip()
```

```python
# Strip any accidental whitespace from column names first
df.columns = df.columns.str.strip()

# Convert date and integer columns
int_cols = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws']
df[int_cols] = df[int_cols].astype(int)

# Convert float columns (indices)
float_cols = ['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']
df[float_cols] = df[float_cols].astype(float)
```

```python
# Create a Region column initialized to 0 (First Region: Bejaia)
df['Region'] = 0

# Assign 1 to the rows after the middle separator (Second Region: Sidi Bel-Abbes)
# Since the split happens exactly at row index 122, we mark everything from 122 onwards as 1
df.loc[122:, 'Region'] = 1

# Convert it to an integer type
df['Region'] = df['Region'].astype(int)
```

```python
print("Dataset shape:",df.shape)
print("-" * 50)
df.info()
```

```python
# 1. Add the Region feature
df['Region'] = 0
df.loc[122:, 'Region'] = 1
df['Region'] = df['Region'].astype(int)

# 2. Drop completely empty rows
df = df.dropna().reset_index(drop=True)

# 3. Drop the redundant text header row from the middle
df = df[df['day'].astype(str).str.strip() != 'day'].reset_index(drop=True)

# 4. Clean up whitespaces from column names and string values
df.columns = df.columns.str.strip()
for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype == 'str':
        df[col] = df[col].str.strip()

# 5. Convert everything to the correct numerical types
int_cols = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Region']
df[int_cols] = df[int_cols].astype(int)

float_cols = ['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']
df[float_cols] = df[float_cols].astype(float)
```

```python
df['Classes'].value_counts()
```

```python
# To transform the text labels into binary number:

# Map 'fire' to 1 and 'not fire' to 0
df['Classes'] = df['Classes'].map({'fire': 1, 'not fire': 0})

print(df['Classes'].value_counts())
```
